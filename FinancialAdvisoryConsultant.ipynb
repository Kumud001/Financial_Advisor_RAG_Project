```python
import tiktoken
import json

#  Use GPT-4's tokenizer for GPT-4o
tokenizer = tiktoken.get_encoding("cl100k_base")  # Same as GPT-4/3.5 tokenizer
# Read JSONL dataset
file_path = "/Users/kumud/Documents/Customer_Support_Finetuning/query_dataset_small.jsonl"
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

# Count tokens in each example
total_tokens = 0
for line in lines:
    data = json.loads(line)
    messages = data["messages"]
    
    # Convert messages to text and tokenize
    text = " ".join([msg["content"] for msg in messages])
    total_tokens += len(tokenizer.encode(text))

print(f"Total tokens in dataset: {total_tokens}")
print(f"Average tokens per example: {total_tokens / len(lines):.2f}")
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Load financial documents
docs = [
    "401(k) accounts offer tax-deferred growth, reducing taxable income.",
    "Roth IRA contributions are taxed upfront but grow tax-free.",
    "A credit score of 750+ is considered excellent for loan approvals.",
    "Mortgage interest rates fluctuate based on credit scores and market trends."
]

# Split into smaller chunks for embedding storage
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_docs = text_splitter.create_documents(docs)

# Convert text into vector embeddings using OpenAI
embeddings = OpenAIEmbeddings(api_key)

# Store embeddings in ChromaDB
vector_db = Chroma.from_documents(split_docs, embedding=embeddings, persist_directory="./chroma_db")

# Save database
vector_db.persist()
```

```python
from langchain.chains import RetrievalQA
from langchain.llms import OpenAI

# Load stored embeddings
vector_db = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)

# Retrieval-based Q&A system
qa_chain = RetrievalQA.from_chain_type(
    llm=OpenAI(api_key),
    retriever=vector_db.as_retriever(),
    chain_type="stuff"
)

def query_rag(question):
    return qa_chain.run(question)

# Test retrieval
print(query_rag("What is the tax advantage of an IRA?"))
```

```python
import fitz # PyMuPDF
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
        return text.strip()
    
    
```

```python
from fastapi import FastAPI, UploadFile, File
from pydantic import BaseModel

app = FastAPI()

class QueryRequest(BaseModel):
    query: str

@app.post("/ask")
def ask_question(request: QueryRequest):
    # Retrieve relevant financial information
    retrieved_info = query_rag(request.query)
    
    # Generate response using fine-tuned GPT
    final_response = query_finetuned_gpt(f"{retrieved_info}\nUser Query: {request.query}")

    return {"response": final_response}

@app.post("/upload_pdf/")
async def upload_pdf(file: UploadFile = File(...)):
    content = await file.read()
    
    # Save and extract text
    with open("uploaded.pdf", "wb") as f:
        f.write(content)
    
    extracted_text = extract_text_from_pdf("uploaded.pdf")
    
    return {"extracted_text": extracted_text}

# Run server: uvicorn filename:app --reload
```
